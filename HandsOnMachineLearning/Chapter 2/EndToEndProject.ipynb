### Looking at the Big Picture
##### Data Metrics:
population, median income, median housing price for each block in California
##### Goal:
Learn from DataSet and predict median housing price in any district/block, given the other metrics

### Frame the Problem
The Problem is a supervised regression task that should be solved via batch learning.

### Select a Performance Measure
A usual measure for regression tasks is the "root mean square error" (RMSE)
$$RMSE(X, h) = \sqrt{\dfrac{1}{m} * \sum \limits _{i=1} ^{m} (h(x^i) - y^i)^2}$$
where "m" is the number of instances in the dataset (2000 Districts means m = 2000)

x^i is a vector of all veature values (excluding the label)of the i-th instance of the dataset

y^i is the label for the i-th instance (the desired output value)
